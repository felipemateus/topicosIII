### 3) 
- Para Primeira questão a regra de aprendizagem:
    $$h_{i+1} = h_i + \alpha(80-4h)$$
    
- Para Segunda questão a regra de aprendizagem:
    $$r_{i+1} = r_i - \alpha(0.30\pi r-37.5\pi r^{-2})$$

### 4)
#### 4.1) Solução da primeira questão

```python
from matplotlib import pyplot as plt
import numpy as np
import math
%matplotlib inline
next_x = 0
gamma = [0.1, 0.05, 0.4, 0.3]
color = ['b','r','g','y']
precision = 0.00001
max_iters = 10000

num_iters = 0

fig,ax = plt.subplots()
def der(x):
    return (80-(4*x))

for j in range(4):
    next_x=0
    current_x = np.zeros(10000)
    num_iters=0
    for i in range(max_iters):
        current_x[i] = next_x
        next_x = current_x[i] + gamma[j]*der(current_x[i])
        step = next_x - current_x[i]
        num_iters+=1
        
        if abs(step)<=precision:
            break
    print("Numero de iterações:"+str(num_iters))
    print("Valor encontrado: "+ str(current_x[num_iters-1]))
    ax.plot(current_x[0:num_iters],color[j],label = 'gamma = '+str(gamma[j]))
    
    
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-small')


plt.xlabel("Iterações")
plt.ylabel("ck")
plt.show()
```

#### 4.1) Solução da segunda questão

```python

from matplotlib import pyplot as plt
import numpy as np
import math
%matplotlib inline
gamma = [0.015, 0.02, 0.05, 0.09]
color = ['b','r','g','y']
precision = 0.01
max_iters = 10000

num_iters = 0

fig,ax = plt.subplots()
def der(x):
    return (0.30*np.pi*x - 37.5*np.pi*(x**(-2)))

for j in range(4):
    next_x=1
    current_x = np.zeros(10000)
    num_iters=0
    for i in range(max_iters):
        current_x[i] = next_x
        next_x = current_x[i] - gamma[j]*der(current_x[i])
        step = next_x - current_x[i]
        num_iters+=1
        
        if abs(step)<=precision:
            break
    print("Numero de iterações:"+str(num_iters))
    print("Valor encontrado: "+ str(current_x[num_iters-1]))
    #ax.axis([0, num_iters,-1,0])
    ax.plot(current_x[0:num_iters],color[j],label = 'gamma = '+str(gamma[j]))
    
    
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-small')

plt.xlabel("Iterações")
plt.ylabel("ck")
plt.show()
```
